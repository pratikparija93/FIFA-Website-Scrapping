### Web Scraper

Website : http://www.fifa.com/fifa-world-ranking/ranking-table/men/

Operations: 
* Collect all of the external links (there must be some on the page of your )
* Associate the link with a textual description of it from the website.
* Write a function to check whether the link is valid.
* Save the external links(urls), textual description, a boolean for valid, and the last vaild datetime check to an excel file.

```python
from __future__ import print_function
import urllib3 #  pip install -i https://pypi.anaconda.org/pypi/simple urllib3
import unicodedata
import requests
from bs4 import BeautifulSoup, SoupStrainer, Comment
import warnings
warnings.filterwarnings('ignore')
```

```python
# To get the response from thr URL Server
url = 'http://www.fifa.com/fifa-world-ranking/ranking-table/men/'
http = urllib3.PoolManager()
response = http.request('GET',url)
response.status
```

```python
page = requests.get('http://www.fifa.com/fifa-world-ranking/ranking-table/men/')
soup = BeautifulSoup(page.text, 'html.parser')
print(soup.prettify())
```

```python
# Collecting all the link on the page
all_link_on_page = soup.find_all('a')
for link in all_link_on_page:
    print(link.get('href'))
```

```python
#Getting the URL and the Text
for social_name in all_link_on_page:
    if social_name.get('href') == None:
        continue
    else:
        links = social_name.get('href')
        names = social_name.text
        print(links)
        print(names)
```

```python
# Import data into Excel File
import xlwt
import time
from datetime import datetime
from tempfile import TemporaryFile
book = xlwt.Workbook()
sheet1 = book.add_sheet('sheet1')
sheet1.write(0,0,'External Links(URLs)')
sheet1.write(0,1,'Textual Description')
sheet1.write(0,2,'Validity')
sheet1.write(0,3,'Timestamp')
i=1
for social_name in all_link_on_page:
    if social_name.get('href') == None:
        continue
    else:
        links = 'http://www.fifa.com/fifa-world-ranking/ranking-table/men/' + social_name.get('href')
        names = social_name.text
        print(links)
        sheet1.write(i,0,links)
        print(names)
        sheet1.write(i,1,names) 
    isvalid = check_valid_link(links)
    print(isvalid)
    sheet1.write(i,2,isvalid)
    ts = time.time()
    q = datetime.fromtimestamp(ts).strftime('%Y-%m-%d %H:%M:%S')
    print(q)
    sheet1.write(i,3,q)
    i+=1
    
filename = "Web_Scrapping.xls"
book.save(filename)
book.save(TemporaryFile())  
print('Excel sheet created')
```
